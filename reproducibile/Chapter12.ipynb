## Avoid Misleading or Confusing Graphs

#### Fox News COVID-19 Daily New Cases

```python
import pandas as pd
import matplotlib.pyplot as plt
dates =  pd.date_range('2020-03-18', periods=15, freq='D')
daily_cases = [33,61,86,112,116,129,192,174,344,304,327,246,320,339,376]
plt.rcParams.update({'font.size': 22})
plt.figure(figsize=(10,6))
plt.plot( dates,daily_cases, label="Daily Cases", marker="o")
plt.legend()
plt.gca().tick_params(axis='x', which='major', labelsize=14)
plt.xlabel("Dates")
plt.ylabel("New Daily Cases")
plt.title("New Daily Cases Versus Dates")
plt.xticks(rotation=45)
for x,y in zip(dates, daily_cases): 
    plt.text(x, y, str(y))
```

### Visualization that confuses audience

```python
from collections import Counter
df = pd.read_excel("PopulationEstimates.xls",skiprows=2)
counter = Counter(df.tail(-1)["Urban_Influence_Code_2003"])
codes = [key for key in counter.keys() if str(key) != "nan"]
heights = [counter[code] for code in codes]
plt.figure(figsize=(10,6))
plt.bar(list(map(lambda x: str(x),codes)), heights)
plt.xticks(rotation=45);
plt.title("Urban Influence Code for All Counties in the US")
plt.xlabel("Urban Influence Code")
plt.ylabel("Count");
```

```python
plt.figure(figsize=(10,6))
plt.gca().set_ylim([80,800])
plt.bar(list(map(lambda x: str(x),codes)), heights)
plt.xticks(rotation=45);
plt.title("Urban Influence Code \nfor All Counties in the US")
plt.xlabel("Urban Influence Code")
plt.ylabel("Count");
```
